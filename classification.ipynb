```python
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import confusion_matrix, f1_score
```

```python
train_path = 'titles-en-train.labeled'
test_path = 'titles-en-test.labeled'

train_sents = []
train_labels = []
with open(train_path) as f:
    for row in f.readlines():
        label, sent = row.strip().split('\t')
        train_sents.append(sent)
        train_labels.append(int(label))

test_sents = []
test_labels = []
with open(test_path) as f:
    for row in f.readlines():
        label, sent = row.strip().split('\t')
        test_sents.append(sent)
        test_labels.append(int(label))
```

```python
tfidf = TfidfVectorizer()
train_X = tfidf.fit_transform(train_sents)
test_X = tfidf.transform(test_sents)
```

```python
# only use tfidf vector as features
# use random forest
rf = RandomForestClassifier()
rf.fit(train_X, train_labels)
pred_labels = rf.predict(test_X)
print(confusion_matrix(test_labels, pred_labels))
print('F1 score : {:.4f}'.format(f1_score(test_labels, pred_labels)))
```
